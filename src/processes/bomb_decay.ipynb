```python
%matplotlib inline
import matplotlib
from matplotlib import pyplot as plt
import numpy as np
```

```python
timestep = np.arange(60, 180060, 300)
expected = np.power((timestep/3600),-1.2)
#expected /= expected[0]
```

```python
offset_factor = np.power((60/3600),-1.2)
print(offset_factor)
stepwise = np.zeros_like(expected)
stepwise[0] = offset_factor # * H+1 == 1 here
totalsteps = 60
for i, ts in enumerate(timestep):
    if i+1 == len(timestep):
        continue
    current = np.power((totalsteps/3600), -1.2)
    nx = np.power(((totalsteps+300)/3600), -1.2)
    stepwise[i+1] = nx/current * stepwise[i]
    totalsteps += 300
    if totalsteps != timestep[i+1]:
        print(totalsteps, timestep[i+1])
```

```python
#plt.xlim(left=0.01)
plt.loglog(timestep/3600, expected)
plt.loglog(timestep/3600, stepwise)
```
