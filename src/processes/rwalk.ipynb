```python

%matplotlib inline
import matplotlib
#matplotlib.use('qtagg')
import matplotlib.pyplot as pl
import math
import numpy

a=0.5 # within boundary, above 0.25
b=0.875
```

```python
vabs = numpy.arange(0,20,.2)

for tstep in [60, 90, 120, 300, 450, 600, 900, 3600]:
    rl = 2*a*((vabs*tstep)**b)
    rl_hour = rl * math.sqrt(3600 / tstep) # error propagation
    pl.plot(vabs, rl_hour, label=tstep)

pl.legend()

```

```python
vabs = numpy.arange(0,20,.2)

for tstep in [60, 90, 120, 300, 450, 600, 900, 3600]:
    rl_hour = 2*a*((vabs*900)**b)
    rl = rl_hour / math.sqrt(900/tstep)
    rl_hourX = rl * math.sqrt(900 / tstep) # error propagation
    pl.plot(vabs, rl_hourX, label=tstep)

pl.legend()
```
